```python
class Ogrenciler:
    def __init__(self, isim, mat, fizik, kimya):
        self.isim = isim
        self.mat = mat
        self.fizik = fizik
        self.kimya = kimya

ogrenci1 = Ogrenciler("Şevval", 90, 70, 80)
ogrenci2 = Ogrenciler("Aylin", 60, 50, 70)
ogrenci3 = Ogrenciler("Sema", 50, 80, 100)
ogrenci4 = Ogrenciler("Gamze", 80, 90, 95)

ogrenciler_listesi = {
    ogrenci1.isim: {"mat": ogrenci1.mat, "fizik": ogrenci1.fizik, "kimya": ogrenci1.kimya},
    ogrenci2.isim: {"mat": ogrenci2.mat, "fizik": ogrenci2.fizik, "kimya": ogrenci2.kimya},
    ogrenci3.isim: {"mat": ogrenci3.mat, "fizik": ogrenci3.fizik, "kimya": ogrenci3.kimya},
    ogrenci4.isim: {"mat": ogrenci4.mat, "fizik": ogrenci4.fizik, "kimya": ogrenci4.kimya},
}

def not_guncelle(ogrenci_ismi, ders, yeni_not):
    if ogrenci_ismi in ogrenciler_listesi:
        if ders in ogrenciler_listesi[ogrenci_ismi]:
            ogrenciler_listesi[ogrenci_ismi][ders] = yeni_not
            print(f"{ogrenci_ismi} adlı öğrencinin {ders} notu güncellendi.")
        else:
            print(f"Geçersiz ders ismi: {ders}")
    else:
        print(f"{ogrenci_ismi} adlı öğrenci bulunamadı.")

def ogrenci_ekle(isim, mat, fizik, kimya):
    if isim not in ogrenciler_listesi:
        ogrenciler_listesi[isim] = {"mat": mat, "fizik": fizik, "kimya": kimya}
        print(f"{isim} adlı öğrenci başarıyla eklendi.")
    else:
        print(f"{isim} adlı öğrenci zaten mevcut.")

def not_goster(ogrenci_ismi, ders):
    if ogrenci_ismi in ogrenciler_listesi:
        if ders in ogrenciler_listesi[ogrenci_ismi]:
            print(f"{ogrenci_ismi} adlı öğrencinin {ders} notu: {ogrenciler_listesi[ogrenci_ismi][ders]}")
        else:
            print(f"Geçersiz ders ismi: {ders}")
    else:
        print(f"{ogrenci_ismi} adlı öğrenci bulunamadı.")

while True:
    print("\nSeçim yapın:")
    print("1 - Not güncelle")
    print("2 - Yeni öğrenci ekle")
    print("3 - Ders notu göster")
    print("4 - Çıkış yap")

    secim = input("Lütfen 1, 2, 3 veya 4 arasında bir seçim yapın: ")

    if secim == "1":
        ogrenci_ismi = input("Öğrenci ismi: ")
        ders = input("Ders ismi (mat, fizik, kimya): ").lower()
        yeni_not = int(input(f"{ogrenci_ismi} adlı öğrencinin yeni {ders} notu: "))
        not_guncelle(ogrenci_ismi, ders, yeni_not)

    elif secim == "2":
        isim = input("Yeni öğrenci ismi: ")
        mat = int(input("Matematik notu: "))
        fizik = int(input("Fizik notu: "))
        kimya = int(input("Kimya notu: "))
        ogrenci_ekle(isim, mat, fizik, kimya)

    elif secim == "3":
        ogrenci_ismi = input("Öğrenci ismi: ")
        ders = input("Ders ismi (mat, fizik, kimya): ").lower()
        not_goster(ogrenci_ismi, ders)

    elif secim == "4":
        print("Çıkış yapılıyor...")
        break

    else:
        print("Geçersiz seçim! Lütfen 1, 2, 3 veya 4 arasında bir seçim yapın.")

     
```
